```r
get_movie_summary <- function(link){
  
  summary_chunks <- glue("https://www.rottentomatoes.com{link}#")
  
    print(summary_chunks)
  
  stopifnot(length(summary_chunks) > 0)
  
  summary <- summary_chunks %>%
     read_html() %>%
    html_nodes(".articleLink.unstyled ") %>% 
    html_text() %>% 
   glue_collapse(sep =  "\n")
  
  return(summary)
}
```
